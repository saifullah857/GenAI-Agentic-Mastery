## Langchain Groq

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
import os

load_dotenv()

model = ChatGroq(
    model="llama-3.1-8b-instant",
    api_key=os.getenv("GROQ_API_KEY")
)

response = model.invoke("How to get internship in AI / ML in pakistan")

print(response.content)
```

## Streaming 

```python
for chunk in model.stream("write a detailed answer in 200 words how to get internship in pakstan in AI / ML "):
    print(chunk.text,end="|",flush=True)
```

```python
for chunk in model.stream("why parrots have colorful feathers ? "):
    print(chunk.text,end="|",flush=True)
```

```python
response = model.invoke("How to get internship in AI / ML in pakistan")

print(response.content)
```

## Batch in Langchain

```python
responses = model.invoke([
"why do parrots have colorful feathrs",
"how do airplan fly",
"what is computer graphics"
])

for response in responses:
    print(response)
```
